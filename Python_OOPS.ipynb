```python
class Car:
    history = []
    def __init__(self, model, year):
        self.model1 = model
        self.year1 = year

    def print_details(self):
        print(f"Details: {self.model1} manufactured in {self.year1}")
        self.history.append((self.model1, self.year1))
    
    @classmethod
    def print_all_cars(self):
        print(self.history)

car1 = Car('CRV',2008)
car1.print_details()
car2 = Car('Hyundai',2010)
car2.print_details()

Car.print_all_cars()

#Inheritance: using Person.__init__() or super.__init__() if a different init is defined
#Poloymorphism: move() can be defined differently in different classes
#Encapsulation: use __age (self.__age) as variable name
```
