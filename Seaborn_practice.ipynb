```python
df = sns.load_dataset('penguins')
```

```python
penguins_df = df.copy()
```

```python
penguins_df['species'].value_counts()
```

```python
penguins_df.groupby('island').agg({
    'island' : 'count'
})
```

```python
sns.scatterplot(data = penguins_df, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex', palette= 'pink')
```

```python
sns.set_style('darkgrid') # darkgrid, whitegrid, white, dark, ticks
sns.scatterplot(data = penguins_df, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex', palette= 'pink')
```

```python
sns.set_style('ticks') # darkgrid, whitegrid, white, dark, ticks
sns.scatterplot(data = penguins_df, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex', palette= 'pink')
```

```python
sns.set_style('ticks') # darkgrid, whitegrid, white, dark, ticks
sns.scatterplot(data = penguins_df, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex', palette= 'pink')
sns.despine(left = True)
```

```python
sns.set_style('ticks') # darkgrid, whitegrid, white, dark, ticks
sns.scatterplot(data = penguins_df, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex', palette= 'pink')
sns.set_context('poster') #paper, notebook (default), poster, talk
```

```python
sns.scatterplot(data = penguins_df, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex', palette= 'coolwarm')
sns.set_context('paper')
```

```python
sns.scatterplot(data = penguins_df, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex', style = 'island')
```
